### 获取音色嵌入

我们只提取目标说话者的音色嵌入。源音色嵌入可以直接从checkpoints_v2/ses文件夹中加载。

### 使用MeloTTS作为基本语音库
这是一个高质量的多语言文本转语音库，由@MyShell.ai开发，支持包括英语（美式、英式、印度式、澳大利亚式、默认）、西班牙语、法语、中文、日语和韩语等多种语言。在下面的示例中，我们将使用MeloTTS中的模型作为基本发声者。



```python
import os
import nltk
import torch

from openvoice import se_extractor
from openvoice.api import ToneColorConverter
from IPython.display import Audio
from melo.api import TTS

ckpt_converter = 'checkpoints_v2/converter'

# 选择设备
device = "cuda:0" if torch.cuda.is_available() else "cpu"

output_dir = 'outputs_v2'

# 初始化音色转换器
tone_color_converter = ToneColorConverter(f'{ckpt_converter}/config.json', device=device)
tone_color_converter.load_ckpt(f'{ckpt_converter}/checkpoint.pth')

# 创建输出目录
os.makedirs(output_dir, exist_ok=True)

# 参考语音文件
reference_speaker = 'resources/Shanghainese.mp3'

# 获取目标语音特征
target_se, audio_name = se_extractor.get_se(reference_speaker, tone_color_converter, vad=True)
# 将目标语音特征移动到指定设备
target_se = target_se.to(device)

# 下载 NLTK 数据
nltk.download('averaged_perceptron_tagger_eng')

# 待合成的文本
texts = {
    'EN': "Did you ever hear a folk tale about a giant turtle?22.Did you ever hear a folk tale about a giant turtle?33.Did you ever hear a folk tale about a giant turtle?",
    'ZH': "11.在这次vacation中，我们计划去Paris欣赏埃菲尔铁塔和卢浮宫的美景。22.在这次vacation中，我们计划去法国欣赏埃菲尔铁塔和卢浮宫。33.在这次假期中，我们计划去法国南部。44.在这次vacation中，我们计划去德国。"
}

# 临时音频文件路径
src_path = f'{output_dir}/tmp.wav'

# 合成速度
speed = 1.0

# 遍历不同语言的文本
for language, text in texts.items():
    # 初始化 TTS 模型
    model = TTS(language=language, device=device)
    # 获取说话人 ID 映射
    speaker_ids = model.hps.data.spk2id

    # 遍历每个说话人
    for speaker_key in speaker_ids.keys():
        # 获取当前说话人的 ID
        speaker_id = speaker_ids[speaker_key]
        # 将说话人 ID 移动到指定设备
        speaker_id = torch.tensor(speaker_id, device=device) if isinstance(speaker_id, int) else speaker_id.to(device)
        speaker_key = speaker_key.lower().replace('_', '-')

        # 加载源语音特征
        source_se = torch.load(f'checkpoints_v2/base_speakers/{language}/{language.lower()}_default_se.pth', map_location=device)
        # 确保源语音特征在指定设备上
        source_se = source_se.to(device)

        # 进行文本转语音合成
        model.tts_to_file(text, speaker_id, src_path, speed=speed)

        # 输出音频文件路径
        save_path = f'{output_dir}/output_v2_{language}.wav'

        # 编码消息
        encode_message = "@MyShell"

        # 进行音色转换
        tone_color_converter.convert(
            audio_src_path=src_path,
            src_se=source_se,
            tgt_se=target_se,
            output_path=save_path,
            message=encode_message
        )
```
